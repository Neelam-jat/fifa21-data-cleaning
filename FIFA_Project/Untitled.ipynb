```python
import pandas as pd

# 1. Load the data
# (This assumes the file is named 'fifa21_raw_data.csv' and is in the same folder)
df = pd.read_csv('fifa21_raw_data.csv', low_memory=False)

# 2. Fix the Height (Convert 5'7" to cm)
def convert_height(x):
    if isinstance(x, str):
        if "'" in x:
            feet, inches = x.split("'")
            inches = inches.replace('"', '')
            return int(feet) * 30.48 + int(inches) * 2.54
        elif 'cm' in x:
            return int(x.replace('cm', ''))
    return x

df['Height_cm'] = df['Height'].apply(convert_height)

# 3. Fix the Weight (Convert 170lbs to kg)
def convert_weight(x):
    if isinstance(x, str):
        if 'lbs' in x:
            return int(x.replace('lbs', '')) * 0.453592
        elif 'kg' in x:
            return int(x.replace('kg', ''))
    return x

df['Weight_kg'] = df['Weight'].apply(convert_weight)

# 4. Show the results
print("Success! Here is a sample of your cleaned data:")
print(df[['Name', 'Height', 'Height_cm', 'Weight', 'Weight_kg']].head())
```
